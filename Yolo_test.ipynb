```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow # Import for Colab compatibility

# 1. 載入預訓練的 YOLOv8 模型
# 這裡使用 'yolov8n.pt' (Nano版本)，它體積最小、速度最快，適合初學者測試
model = YOLO('yolov8n.pt')

# 2. 指定要偵測的圖片路徑 (請替換為您電腦中的圖片路徑)
#image_path = 'https://ultralytics.com/images/bus.jpg'
image_path = 'test.jpg'
# 3. 執行物件偵測預測
# save=True：會自動將畫好標記框的結果圖片存檔（預設存放在 runs/detect/predict 目錄下）
# conf=0.5：信心閥值，代表只標示出模型有 50% 以上把握的物件
results = model.predict(source=image_path, save=True, conf=0.5)

# 4. (選擇性) 透過 OpenCV 在螢幕上即時顯示標註後的圖片
# results[0].plot() 會將帶有標記框的圖片轉換為 numpy 陣列格式
annotated_frame = results[0].plot()

# 顯示圖片
cv2_imshow(annotated_frame) # Use cv2_imshow for Colab

# cv2.waitKey(0) and cv2.destroyAllWindows() are not needed with cv2_imshow in Colab
```
